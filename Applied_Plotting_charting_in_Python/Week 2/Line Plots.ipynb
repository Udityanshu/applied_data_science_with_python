# Line Plots

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2
plt.figure()

# plot the linear data and exponential data
plt.plot(linear_data, '-o', exponential_data, '-o')
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2
plt.figure()

# plot the linear data and exponential data
plt.plot(linear_data, '-o', exponential_data, '-o')

#plot another series with a dashed red line
plt.plot([22,35,44], '--r')
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2
plt.figure()

# plot the linear data and exponential data
plt.plot(linear_data, '-o', exponential_data, '-o')

#plot another series with a dashed red line
plt.plot([22,35,44], '--r')

plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('Relationship between Some data and Some other data')
# add a legend with legend entries (because we didn't have labels when we plotted the data series)
plt.legend(['Baseline', 'Competition', 'Us'])
```

```python
linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2
plt.figure()

# plot the linear data and exponential data
plt.plot(linear_data, '-o', exponential_data, '-o')

#plot another series with a dashed red line
plt.plot([22,35,44], '--r')

plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('Relationship between Some data and Some other data')
# add a legend with legend entries (because we didn't have labels when we plotted the data series)
plt.legend(['Baseline', 'Competition', 'Us'])

# fill the area between the linear data and exponential data
plt.gca().fill_between(range(len(linear_data)),
                      linear_data, exponential_data,
                      facecolor='blue',
                      alpha=0.25)
```

Trying working with dates

```python
plt.figure()
observation_dates = np.arange('2019-01-01', '2019-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',
         observation_dates, exponential_data, '-o')
```

```python
plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # convert the map to a list to get rid of the error
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')
```

rotate the tick labels for x axis

```python
plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # convert the map to a list to get rid of the error
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)
```

adjust the subplot so the text doesn't run off the image

```python
plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # convert the map to a list to get rid of the error
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)

# adjust the subplot so the text doesn't run off the image
plt.subplots_adjust(bottom=0.15)
```

```python
plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # convert the map to a list to get rid of the error
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)

# adjust the subplot so the text doesn't run off the image
plt.subplots_adjust(bottom=0.15)

# setting up the xaxis and yaxis labels following with the chart title
ax = plt.gca()
ax.set_xlabel('Dates')
ax.set_ylabel('Units')
ax.set_title('Exponential vs Linear performance')
```

```python
plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) # convert the map to a list to get rid of the error
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)

# adjust the subplot so the text doesn't run off the image
plt.subplots_adjust(bottom=0.15)

# setting up the xaxis and yaxis labels following with the chart title
ax = plt.gca()
ax.set_xlabel('Dates')
ax.set_ylabel('Units')
ax.set_title('Exponential vs Linear performance')

# adding mathematical expressions in any text element
ax.set_title('Exponential ($x^2$) vs Linear Expression ($x$)')
```
